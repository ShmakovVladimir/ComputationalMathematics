__Упражнение 1__



$$
x = k \cdot sin(t)^{3} \\
y = 13 cos(t) - 5 cos(2t) -2cos(3t)-cos(4t) \\
t \in [0, 4 \pi]
$$

```python
fig, ax = plt.subplots(2, 2, figsize = (8, 8), dpi = 200)

k = np.array([[10, 1],[16, 1000]])
t = np.linspace(0, 4 * np.pi ,100)
colors = [['#ff6885', '#e9ba46'], ['#df0d0d', '#cf004b']]
y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3  * t) - np.cos(4 * t)
for i in range(2): 
    for j in range(2):
        x = np.power(np.sin(t), 3) * k[i, j]
        ax[i, j].plot(x, y, color = colors[i][j])
        ax[i,j].grid()
        ax[i, j].set_title(f'k = {k[i, j]}', fontsize = 16)
```

Построить график с крестами ошибок

```python
i = [0, 5, 10, 20, 32] #сила тока
u = np.array([0, 10, 20, 40, 66]) #напряжение на резисторе
u_err, i_err = 1, 2 #ошибка определения тока и напряжения

plt.figure(figsize = (5, 5), dpi = 200)
plt.errorbar(i, u, yerr = [u_err for _ in u], color = 'red', xerr = [i_err for _ in i])
plt.xlabel("Сила тока $I [А]$", fontsize = 12)
plt.ylabel("Напряжение $U [В]$", fontsize = 12)
plt.grid()
plt.title("Зависимость напряжения от силы тока", fontsize = 12)
```
