```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
from matplotlib import cm
```

```python
def random_values(cov: np.ndarray, length = 1000):
    A = np.linalg.cholesky(cov)
    z = sts.norm(loc = 1, scale = 1).rvs(size = (cov.shape[0], length))
    m = np.ones_like(z)
    x = m + A @ z
    return x
    
```

```python

N = np.random.randint(2, 10)
cov = np.random.rand(N, N)
cov = cov @ cov.T 
x = random_values(cov)
print("Сгенерированная ковариационная матрица")
print(cov)
print("Оценка ковариационной матрицы")
print(np.cov(x))
```

```python
r = np.array([2.7, 3.7, 4.9, 5.6, 6.3, 7.1, 7.5])
m = np.arange(len(r)) + 1
```

```python
plt.figure(figsize = (12, 5), dpi = 200)
mnk = sts.linregress(m, r)
x = np.linspace(1, 7, 500)
plt.plot(x, mnk.slope * x + mnk.intercept)
plt.scatter(m, r, label = 'Экспериментальные данные')

plt.legend(loc = 'upper left')
plt.xlabel("$m$", fontsize = 16)
plt.ylabel(r"$r^{2} \ [см^{2}]$", fontsize = 16)
plt.title("Зависимость квадрата радиуса кольца от порядка максимума", fontsize = 16)
plt.grid()
```

```python
x = np.linspace(0, 5, 30)
y = 5 * x + sts.norm(loc = 0, scale = 1).rvs(x.shape)
x_line = np.linspace(0, 6, 1000)
MNK = sts.linregress(x, y)

plt.figure(figsize = (12, 5), dpi = 200)
plt.scatter(x, y, color = 'purple')

plt.plot(x_line, 
         MNK.intercept + MNK.slope * x_line, 
         color = 'green')
plt.grid(ls = ':')
plt.xticks(np.arange(0, 5.5, 0.5))
plt.xlim(0, 5);
```
